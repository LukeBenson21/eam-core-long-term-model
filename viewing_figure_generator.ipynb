# Viewing Figures

This Notebook is inteaded to tranform the data to fit the model parameters

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
views_per_hour = "./iplayer_views_per_hour.xlsx"
views_per_hour_df = pd.read_excel(views_per_hour)
#views_per_hour_df = views_per_hour_df.set_index("Hour")
views_per_hour_df = views_per_hour_df.drop(columns=["Hour"])
views_per_hour_df
```

covert to fractions

```python
def convet_to_fraction(cell):
    #print(cell)
    #print(str(cell))
    array = str(cell).split(":")
    hour = array[0]
    minutes = array[1]
    seconds= array[2]
    hour = int(hour) * 60
    seconds = float(seconds)/60
    minutes_fraction = int(hour)+int(minutes)+float(seconds)
    return minutes_fraction

views_per_hour_df = views_per_hour_df.applymap(convet_to_fraction)
views_per_hour_df
```

Need to convert to 365 days for 0-23 Hours a day. Applying the months to the suitable days. Base of 2022

```python
jan_index = 31
feb_index = jan_index + 28
mar_index = feb_index + 31
apr_index = mar_index + 30
may_index = apr_index + 31
jun_index = may_index + 30
jul_index = jun_index + 31
aug_index = jul_index + 31
sep_index = aug_index + 30
oct_index = sep_index + 31
nov_index = oct_index + 30
dec_index = nov_index + 31

month_index = {jan_index:"January",feb_index:"February",mar_index:"March",apr_index:"April",may_index:"May",jun_index:"June",jul_index:"July",aug_index:"August",sep_index:"September",oct_index:"October",nov_index:"November",dec_index:"December"}

def get_referance_month(day_num, month_index):
    for index in month_index:
        index = index
        if day_num < index:
            return month_index[index]
            break
```

```python
output_df = pd.DataFrame()
output_df["group"] = []
output_df["ref value"] = []
output_df["day"]=[]
output_df["hour"]=[]

for day_num in range(0,365):
    reference_month = get_referance_month(day_num,month_index)
    reference_times = views_per_hour_df[reference_month]
    for hour_num in range(0,24):
        datetime_value = "d_"+str(day_num).zfill(3)+"_h_"+str(hour_num).zfill(2)
        time = reference_times[hour_num]
        row = {"group":datetime_value,"ref value": time,"day":day_num,"hour":hour_num}
        output_df = output_df.append(row,ignore_index=True)
```

```python
output_df = output_df.set_index("group")
output_df = output_df.drop(columns=["day","hour"])
output_df
```
